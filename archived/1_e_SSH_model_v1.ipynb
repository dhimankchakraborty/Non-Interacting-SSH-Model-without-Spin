```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normalize(v):
    norm = np.linalg.norm(v)
    if norm == 0: 
       return v
    return v / norm
```

```python
v = 0.5
w = 1
N = 10
N_site = N * 2

E_arr = np.zeros((N_site), dtype=np.float64)
E_vec_arr = np.zeros((N_site, N_site), dtype=np.float64)

#populate H:
H = np.zeros(( 2*N,2*N ),dtype=np.float64)
for k in range(N):
  H[2*k-1,2*k] = w
  H[2*k,2*k-1] = w
  H[2*k,2*k+1] = v
  H[2*k+1,2*k] = v
H[-1,0] = 0
H[0,-1] = 0

print(H)
# print(np.shape(H))
e_val_arr, e_vec_arr = np.linalg.eigh(H)
idx = e_val_arr.argsort()
E_arr = e_val_arr[idx]
E_vec_arr = e_vec_arr[:, idx]
```

```python
for i, energy in enumerate(E_arr):
    print(f"{i} ---->> {energy}")
print(np.min(np.abs(E_arr)))
print(np.argmin(np.abs(E_arr)))

e_vec_normalized = normalize(E_vec_arr[np.argmin(np.abs(E_arr))])
e_vec_normalized_2 = normalize(E_vec_arr[10])

pos_arr = np.arange(1, N_site + 1)

plt.plot(pos_arr, e_vec_normalized, marker='o')
plt.grid()
plt.show()
plt.plot(pos_arr, e_vec_normalized_2, marker='o')
plt.grid()
plt.show()
```
